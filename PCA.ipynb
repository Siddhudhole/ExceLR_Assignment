### Model fit on original data 

```python
from sklearn.cluster import KMeans 
```

```python
sse = [] #SUM OF SQUARED ERROR
for k in range(1,11): 
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(df)
    sse.append(km.inertia_) 
```

```python
sns.set_style("whitegrid")
g=sns.lineplot(x=range(1,11), y=sse)
 
g.set(xlabel ="Number of cluster (k)", 
      ylabel = "Sum Squared Error", 
      title ='Elbow Method')
 
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(df) 
```

```python
labels = model.predict(df) 
```

### Silhouette score 

```python
from sklearn.metrics import silhouette_score 
```

```python
print('silhouette score :',silhouette_score(df,labels=labels))
```

### Plot Clusters 

```python
from sklearn.decomposition import PCA 

PCA_model = PCA(n_components=2)

scaled = PCA_model.fit_transform(df) 
```

```python
sns.scatterplot(scaled)
```

### Feature Scaling  

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()
```

```python
data_scaled = scaler.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(data=data_scaled,columns=scaler.get_feature_names_out())
```

```python
df_scaled.head()
```

### PCA 

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(7) 
pca.fit(df_scaled)
```

```python
sum(pca.explained_variance_ratio_)
```

```python
explained_variance = pca.explained_variance_ratio_ 
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```
