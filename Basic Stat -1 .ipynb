```python
# Standardization of Numerical Variables 
from sklearn.preprocessing import StandardScaler 
from sklearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder 
```

```python
num_df = df.select_dtypes(exclude='O')
```

```python
num_df.columns
```

```python
scaler = StandardScaler() 
```

```python
num_tras = ColumnTransformer(transformers=[('scalr',StandardScaler(),[0,1,2,3,4])],remainder='passthrough') 
```

```python
num_tras_df = pd.DataFrame(num_tras.fit_transform(num_df),columns=list(num_df.columns)) 
```

```python
num_tras_df.sample(5)
```

```python
cat_df = df.select_dtypes(include='O')
```

```python
cat_df.head() 
```

```python
cat_tras = ColumnTransformer(transformers=[('encode',OneHotEncoder(drop='first',sparse_output=False,dtype='int'),[1,2,3,4,5])])
```

```python
cat_tras_df = pd.DataFrame(cat_tras.fit_transform(cat_df),columns=cat_tras.get_feature_names_out())
```

```python
cat_tras_df.sample(5)
```

```python
final_df = pd.concat([num_tras_df,cat_tras_df],axis=1)
```

```python
final_df.sample(5) 
```
